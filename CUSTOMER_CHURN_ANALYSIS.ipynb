```python
import pandas as pd

url = "https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv"
df = pd.read_csv(url)
```

```python
df['Churn'].unique()
```

```python
df['Churn'] = df['Churn'].str.strip()
```

```python
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
df['Churn'].value_counts()
```

```python
import matplotlib.pyplot as plt                                             # 1 = CUSTOMER CHURNED ( CUSTOMER LEFT THE SERVICE)
import seaborn as sns                                                        # 0 = CUSTOMER STAYED

sns.countplot(x='Churn', data=df)
plt.title("Customer Churn Distribution")
plt.show()
```

```python
sns.boxplot(x='Churn', y='tenure', data=df)
plt.title("Tenure vs Churn")
plt.show()
```

```python
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Churn by Contract Type")
plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

```python
sns.countplot(x='TechSupport', hue='Churn', data=df)
plt.title("Churn by Tech Support")
plt.show()
```

```python
# The company has a churn rate of approximately 26%,
 # which represents a significant risk to recurring revenue and long-term growth.
 # Analysis indicates that churn is primarily driven by the following factors:
   #                   customers with low tenure are more likely to leave during the early stages of their lifecycle;
    #                  customers on month-to-month contracts exhibit significantly higher churn due to lack of long-term commitment;
     #                 customers with higher monthly charges show increased churn, indicating price sensitivity; and
      #                customers without access to tech support are more likely to churn, highlighting the importance of support services in retention.
```
